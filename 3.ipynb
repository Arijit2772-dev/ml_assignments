```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA
```

1) part

```python
column_names = [
    "symboling", "normalized_losses", "make", "fuel_type", "aspiration",
    "num_doors", "body_style", "drive_wheels", "engine_location", "wheel_base",
    "length", "width", "height", "curb_weight", "engine_type", "num_cylinders",
    "engine_size", "fuel_system", "bore", "stroke", "compression_ratio",
    "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
]
```

```python
df_cars = pd.read_csv("imports-85.csv", header=None, names=column_names, na_values="?")
print("Step 1: Dataset loaded from 'imports-85.csv'.")
```

```python
df_cars.head()
```

2) part

```python
df_cars.dropna(subset=['price'], inplace=True)

for col in df_cars.columns:
    if df_cars[col].dtype == 'object':
        df_cars[col].fillna(df_cars[col].mode()[0], inplace=True)
    else:
        df_cars[col].fillna(df_cars[col].median(), inplace=True)
```

```python
df_cars.head()
```

3) part 

```python
num_map = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
df_cars['num_doors'] = df_cars['num_doors'].replace(num_map)
df_cars['num_cylinders'] = df_cars['num_cylinders'].replace(num_map)
```

```python
df_cars.head()
```

```python
df_cars = pd.get_dummies(df_cars, columns=['body_style', 'drive_wheels'], drop_first=True)

label_cols = ['make', 'aspiration', 'engine_location', 'fuel_type']
```

```python
for col in label_cols:
    le = LabelEncoder()
    df_cars[col] = le.fit_transform(df_cars[col])
```

```python
df_cars['fuel_system'] = np.where(df_cars['fuel_system'].str.contains('pfi'), 1, 0)
df_cars['engine_type'] = np.where(df_cars['engine_type'].str.contains('ohc'), 1, 0)
```

```python
df_cars.head()
```

4) part

```python
y = df_cars['price']
X = df_cars.drop('price', axis=1)
X = X.apply(pd.to_numeric) 

scaler_q3 = StandardScaler()
X_scaled = scaler_q3.fit_transform(X)
print("Step 4: Data divided and scaled.")
```

5) part

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

model_full = LinearRegression()
model_full.fit(X_train, y_train)
score_full = model_full.score(X_test, y_test)
print(f"\nStep 5: Model performance on ALL features: R2 Score = {score_full:.4f}")
```

6) part

```python
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print(f"\nOriginal number of features: {X_train.shape[1]}")
print(f"Number of features after PCA: {X_train_pca.shape[1]}")

model_pca = LinearRegression()
model_pca.fit(X_train_pca, y_train)
score_pca = model_pca.score(X_test_pca, y_test)
print(f"Step 6: Model performance on PCA-reduced features: R2 Score = {score_pca:.4f}")

if score_pca > score_full:
    print("\nConclusion: Yes, PCA led to a slight performance improvement!")
else:
    print("\nConclusion: No, PCA did not improve performance, but it simplified the model.")
```
